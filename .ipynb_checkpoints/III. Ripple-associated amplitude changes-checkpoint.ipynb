# Ripple-associated amplitude changes

The forth notebook is used to analyse putative ripple-associated CG amplitude changes (that is, during hippocampal ripple occurrence) in different task contexts. It does so by:

- Plotting the averaged amplitude (uV) before and after ripple occurrence during different task phases or contexts;
- Calculating the baselined voltage before and after ripple occurrence
- Plotting amplitude for all SWR events in a heatmap;
- Significance testing of observed changes to baseline. PENDING

## Comparing ERP windows

Pre-onset period baseline to  previous baseline period average
Post onset baselined to previous period average.

```python
# Label intervals
data.loc[(data.bin_centres >= -0.40) & (data.bin_centres < -0.1), 'timepoint'] = 'Baseline'
data.loc[(data.bin_centres >= -0.1) & (data.bin_centres < 0), 'timepoint'] = 'Pre'
data.loc[(data.bin_centres > 0) & (data.bin_centres <= 0.1), 'timepoint'] = 'Post'
```

```python
# Calculate baseline / pre averages per ripple
baseline_averages = data[data.timepoint=='Baseline'].groupby(['ripple_nr'])['voltage'].mean().reset_index()
pre_averages = data[data.timepoint=='Pre'].groupby(['ripple_nr'])['voltage'].mean().reset_index()
```

```python
# Calculate baseline / pre averages per ripple
baseline_sd = data[data.timepoint=='Baseline'].groupby(['ripple_nr'])['voltage'].std().reset_index()
pre_sd = data[data.timepoint=='Pre'].groupby(['ripple_nr'])['voltage'].std().reset_index()
```

```python
# Subtract baseline averages to pre- and post- period
dict_renaming = {'voltage_x':'voltage', 'voltage_y':'avg', 'voltage':'sd'}
pre = data[data.timepoint=='Pre'].merge(
    baseline_averages, on='ripple_nr', how='left').merge(
    baseline_sd, on='ripple_nr', how='left').rename(
    columns=dict_renaming
    )

post = data[data.timepoint=='Post'].merge(
    baseline_averages, on='ripple_nr', how='left').merge(
    baseline_sd, on='ripple_nr', how='left').rename(
    columns=dict_renaming
)

pre['voltage_baselined'] = (pre['voltage'] - pre['avg'])
post['voltage_baselined'] = (post['voltage']- post['avg'])
```

## Statistical testing

### Test normality function

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def test_normality(data):
    # Ensure the data is a pandas Series
    if not isinstance(data, pd.Series):
        raise TypeError("Input data must be a pandas Series.")

    # Drop NA values to avoid issues with statistical tests
    data = data.dropna()

    # Shapiro-Wilk Test
    shapiro_test = stats.shapiro(data)
    print("Shapiro-Wilk Test:")
    print(f"Statistic: {shapiro_test.statistic}, p-value: {shapiro_test.pvalue}")

    # Kolmogorov-Smirnov Test
    ks_test = stats.kstest(data, 'norm', args=(data.mean(), data.std()))
    print("\nKolmogorov-Smirnov Test:")
    print(f"Statistic: {ks_test.statistic}, p-value: {ks_test.pvalue}")

    # Visual Methods
    plt.figure(figsize=(12, 6))

    # Histogram with KDE
    plt.subplot(1, 2, 1)
    sns.histplot(data, kde=True)
    plt.title('Histogram with KDE')

    # Q-Q Plot
    plt.subplot(1, 2, 2)
    stats.probplot(data, dist="norm", plot=plt)
    plt.title('Q-Q Plot')

    plt.tight_layout()
    plt.show()
```

### Normality check

```python
## Test Normality
for phase_label in to_analyze.phase.unique():
    for t in to_analyze.timepoint.unique():
        print(phase_label, t)
        test_normality(np.log10(to_analyze.loc[(to_analyze.phase==phase_label) & (to_analyze.timepoint==t), 'voltage']))
```
