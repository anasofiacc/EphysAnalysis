```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
perf = pd.Series([75,80,80,80,80,85,85,85,93,95], name='performance')
```

```python
plt.Figure(dpi=300, figsize=(3,3))
sns.set(style='white', context='talk')

performances = [75,80,80,80,80,85,85,85,93,95]


sns.barplot(
    data=performances,
    estimator=np.mean,
    errorbar='sd',
    color='lightgray',
    capsize=.03,
    errwidth=1.5,
    width=.1)

plt.ylim([0,100])
plt.xlim([-.1,.1])
# Remove x-tick labels
plt.xticks([])
plt.ylabel('Performance (%)')
sns.despine()
```
